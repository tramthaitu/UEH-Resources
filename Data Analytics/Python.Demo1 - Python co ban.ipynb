```python
## Hiển thị giá trị kiểu số (cách đơn giản)
nb_float  = 2.123496789
nb_int    = 200
## '%xd': thụt khoảng trống vào x ký tự
print(' a)', '%8d' %nb_int) ## độ dài 8
print(' b)', '{:4d}'.format(nb_int))

print('c1)', '%08d' %nb_int) ## thay space phía trước = 0
print('c2)', '{:08d}'.format(nb_int))

## '%.xf': làm tròn x chữ số thập phân
print(' d)', '%.4f' %nb_float) ## 4 chữ số thập phân

print(' e)', '{:f}'.format(nb_float)) ## ngầm định 6 chữ số thập phân
print('f1)', '{:10.4f}'.format(nb_float)) ## độ dài 10, 4 chữ số thập phân
print('f2)', '%*.*f' % (10, 4, nb_float)) ## độ dài 10, 4 chữ số thập phân
print('g1)', '%+d' %nb_int) ## có dấu +
print('g2)', '{:+d}'.format(nb_int)) ## có dấu +
print('h1)', '%d' % (-1 * nb_int)) ## có dấu -
print('h2)', '{:d}'.format(-1 * nb_int)) ## có dấu -
print(' i)', '{:=5d}'.format((-1 * nb_int))) ## thêm space giữa dấu và số
## Tìm hiểu thêm: print(f'{:}')
```

```python
## Hiển thị giá trị kiểu chuỗi
s1, s2 = 'Nước Việt Nam', 'Hà Nội'
### Canh TRÁI
print('a1)', '%-20s, %-20s' %(s1, s2)) ## độ dài 20
print('a2)', '{:20}, {:20}'.format(s1, s2)) ## độ dài 20
print(' b)', '{:_<20}, {:20}'.format(s1, s2)) ## thay space bằng _
### Canh PHẢI
print('c1)', '%20s, %20s' %(s1, s2)) ## độ dài 20
print('c2)', '{:>20}, {:>20}'.format(s1, s2)) ## độ dài 20
print(' d)', '{:_>20}, {:>20}'.format(s1, s2)) ## thay space bằng _
### Canh GIỮA
print(' e)', '{:^20}, {:^20}'.format(s1, s2)) ## độ dài 20
### Phối hợp TRÁI, PHẢI
print('f1)', '%-20s, %20s' %(s1, s2)) ## độ dài 20
print('f2)', '{:20}, {:>20}'.format(s1, s2)) ## độ dài 20
### Cắt chiều dài chuỗi
print('g1)', '%.8s' % (s1)) ## độ dài 8
print('g2)', '{:.8}'.format(s1)) ## độ dài 8
```

#### **4.2 Hàm (function)**

```python
## Hàm tính giai thừa bằng vòng lặp
def giai_thua_for(n):
    giai_thua = 1
    for i in range(1, n + 1):
        giai_thua *= i

    return giai_thua
```

```python
## Gọi hàm
n = int(input('Nhập số nguyên (n < 10) '))
print('Sử dụng vòng lặp:', str(n) + '! =', giai_thua_for(n))
```

```python
## Hàm có thể gọi thực hiện chính nó: ĐỆ QUY
def giai_thua_de_quy(n):
    if(n <= 1):
        return 1

    return(n * giai_thua_de_quy(n - 1))
```

```python
## Gọi hàm
print('Sử dụng đệ quy :', str(n) + '! =', giai_thua_de_quy(n))
```

```python
## Hàm có thể trả về nhiều kết quả
import math
PI = math.pi
def hinh_tron(ban_kinh):
    chu_vi    = (ban_kinh * 2) * PI
    dien_tich = (ban_kinh ** 2) * PI
    return chu_vi, dien_tich
```

```python
## Gọi hàm
r      = eval(input('Cho biết bán kính '))
cv, dt = hinh_tron(r)
print(f'Chu vi hình tròn = {cv:.2f}, diện tích hình tròn = {dt:.2f}')
```

```python
## Kết quả có thể có kiểu dữ liệu tùy ý
def tieu_de(s, w):
    kq = [s.ljust(w), s.center(w), s.rjust(w)] ## canh biên: trái, giữa, phải
    return kq
```

```python
chuoi = tieu_de('Cộng Hòa Xã Hội Chủ Nghĩa Việt Nam', 50) ## Canh biên tiêu đề với độ rộng 50
print('Tiêu đề trái: ' + chuoi[0])
print('Tiêu đề phải: ' + chuoi[1])
print('Tiêu đề giữa: ' + chuoi[2])
```

```python
## Giá trị ngầm định cho tham số
def hinh_tron(ban_kinh, PI = 3.14):
    chu_vi    = (ban_kinh * 2) * PI
    dien_tich = (ban_kinh ** 2) * PI

    return chu_vi, dien_tich
```

```python
## Truyền tham số theo GIÁ TRỊ (by VALUE)
def by_value(param1, param2):
    param1 *= 2 #Nhân đôi và gán lại cho chính nó
    result  = param1 * param2

    return result
```

```python
## Gọi thực hiện
x, y  = 2, 8
value = by_value(x, y)
print(f'x = {x}')
```

```python
## Truyền tham số theo ĐỐI TƯỢNG THAM CHIẾU (by OBJECT REFERENCE)
def by_reference(param1, param2):
    param1[0] = 27
    result    = param1 * param2

    return result
```

```python
## Gọi thực hiện
lst, y  = [1, 2, 3], 8
value = by_reference(lst, y)
print(f'List = {lst}')
```

```python
## Sử dụng *args (to pass a non-keyworded, variable-length argument list)
def func1(*args):
    for arg in args:
        print (arg)
```

```python
## Gọi thực hiện với tham số *argv
func1('Python', 'Programming', 'Language')
func1('Python', 'Language')
```

```python
## Sử dụng **kwargs (to pass a keyworded, variable-length argument list)
def func2(**kwargs):
    for kw, value in kwargs.items():
        print (f'{kw} = {value}')
```

#### **4.3 Hàm lambda (anonymous function)**
---

```python
## Hàm lambda đơn giản
dien_tich = lambda chieu_dai, chieu_rong: (chieu_dai * chieu_rong)
print(f'Diện tích hình chữ nhật = {dien_tich(15, 2)}')
```

```python
## Hàm lambda với điều kiện if-else
nb = lambda num: f'{num}' if num >= 0 else f'-{abs(num)}'

print("Số không âm:", nb(100))
print("Số âm      :", nb(-23))
```

```python
def lookandsay(number):
    result = ""

    repeat = number[0]
    number = number[1:]+" "
    times = 1

    for actual in number:
        if actual != repeat:
            result += str(times)+repeat
            times = 1
            repeat = actual
        else:
            times += 1

    return result

num = "3"

for i in range(10):
    print (num)
    num = lookandsay(num)
```

```python
## Sử dụng để duyệt list
even_numbers = [lambda arg = x: arg * 10 for x in range(1, 10)]
for e in even_numbers:
    print(e())
```
