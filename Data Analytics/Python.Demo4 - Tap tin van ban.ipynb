### **Xử lý tập tin văn bản (.TXT)**
---

* **Tập tin văn bản (Text file):**
  * Tập tin .txt
  * Không có cấu trúc, dãy liên tục các ký tự trên 1 hay nhiều dòng
  * \n ở cuối mỗi dòng (ẩn)
  * **BOF (Begin of file):** Góc bên trái trên cùng
  * **EOF (End of file):** Góc bên phải dưới cùng
  * **pointer:** con trỏ để đọc ghi vị trí hiện hành, sau khi đọc ghi nó sẽ tự động di chuyển đến vị trí kế tiếp

* **Khai thác tập tin văn bản:**
  * mở, đóng tập tin
  * truy cập (đọc/ghi) dựa trên vị trí hiện hành của con trỏ (**pointer**)
  * đọc (**read**): Theo 2 cơ chế: **theo ký tự** & **theo dòng**
  * ghi (**write**): Theo 2 cơ chế: **theo ký tự** & **theo dòng**

* ***Mở file*** ⇒ Chép file từ bộ nhớ thứ cấp lên RAM
* ***Đóng file*** ⇒ Lưu dữ liệu từ RAM xuống đĩa cứng



*   **r:** Mở để đọc, mode mặc định
*   **r+:** Mở để đọc và ghi
*   **w:** Mở để ghi. Trước đó, nó sẽ xóa hết nội dung của file hiện có.
    Nếu file không tồn tại, sẽ tạo ra một file với tên là tên file chúng ta truyền vào.
*   **w+:** Mở để ghi và đọc. Trước đó cũng đã xóa hết nội dung của file hiện có.
    Nếu file không tồn tại, sẽ tạo ra một file với tên là tên file chúng ta truyền vào.
*   **a**: Mở để ghi. Nếu file không tồn tại, sẽ tạo ra một file với tên là tên file chúng ta truyền vào
*   **a+**: Mở để ghi và đọc. Nếu file không tồn tại, sẽ tạo ra một file với tên là tên file chúng ta truyền vào.




#### *Mở và xem nội dung tập tin .txt*

```python
##------------------------------------------------------------------------------
## Một số hàm, phương thức
##   - open( ), close( ): mở, đóng file (built-in functions)
##   - tell( ): vị trí con trỏ hiện hành
##   - seek( ): di chuyển con trỏ
##   - read( ): đọc theo ký tự
##   - readline( ): đọc theo dòng
##   - readlines( ): đọc tất cả các dòng và đưa vào list
##   - write( ): ghi theo ký tự
##   - writelines( ): ghi list
##------------------------------------------------------------------------------
```

```python
## File Van_ban.txt được mở chỉ để ĐỌC
txt_file = open('/content/gdrive/My Drive/Colab Notebooks/DA/Classnote/Data/Van_ban.txt', 'r')
print(f'* Vị trí ban đầu: {txt_file.tell()}') # đầu file (0: begin of file)

## Nội dung của file
content = txt_file.readlines()
content = ''.join(content) # chuyển từ list sang string
print(f'* Nội dung của file:\n{content}')
print(f'* Số dòng: {len(content)}')
print(f'* Vị trí sau khi đọc: {txt_file.tell()}') # cuối file (end of file)

txt_file.close()
```

```python
## Đọc nội dung file .txt theo ký tự
txt_file = open('/content/gdrive/My Drive/Colab Notebooks/DA/Classnote/Data/Van_ban.txt', 'r')

text = txt_file.read(6) # đọc 6 ký tự từ vị trí hiện hành
print(text)

txt_file.seek(20) # dịch chuyển đến vị trí 20
text = txt_file.read(17) # đọc 17 ký tự từ vị trí hiện hành
print(text)

txt_file.close()
```

```python
## Đọc nội dung file .txt theo dòng
txt_file = open('/content/gdrive/My Drive/Colab Notebooks/DA/Classnote/Data/Van_ban.txt', 'r')

line = txt_file.readline() # dòng hiện hành: nội dung (bỏ ký tự cuối cùng: \n)
print('Nội dung: \'' + line[:-1] + '\'', '--> Con trỏ:', txt_file.tell())

line = txt_file.readline()
print('Nội dung: \'' + line[:-1] + '\'', '--> Con trỏ:', txt_file.tell())

line = txt_file.readline() # dòng cuối không cần bỏ ký tự cuối cùng)
print('Nội dung: \'' + line + '\'', '--> Con trỏ:', txt_file.tell())

txt_file.close()
```

```python
## Duyệt các dòng trong file
txt_file = open('/content/gdrive/My Drive/Colab Notebooks/DA/Classnote/Data/Van_ban.txt', 'r')

for line in txt_file:
    print(line) # line có chứa \n ở cuối dòng
txt_file.close()
```

```python
## File Van_ban.txt được mở để GHI thêm vào cuối file (append) và ĐỌC
txt_file = open('/content/gdrive/My Drive/Colab Notebooks/DA/Classnote/Data/Van_ban.txt', 'a+')
txt_file.write('\nDong moi them thu 1')
txt_file.write('\nDong moi them thu 2')
txt_file.seek(0)
print(txt_file.read())

txt_file.close()
```
