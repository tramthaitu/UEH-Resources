### **1. Series**
---


*   Mảng một chiều
*   Có thể tự định nghĩa index của phần tử

##### *Tạo series chứa các phần tử*

```python
## Sử dụng chỉ số (ngầm định: 0, 1, 2, ...)
s = pd.Series([9, 2, 7, 1, 6])  #Xem như một mảng một chiều, dưới góc nhìn của pandas -> series
print('Số phần tử :', s.size)

print(s, '\n')

print('Các CHỈ SỐ :', s.index) #Có thể tự định nghĩa các chỉ số cho các phần tử, hoặc xài ngầm định
print('Các CHỈ SỐ :', s.axes)
print('Các GIÁ TRỊ:', s.values) # kiểu array, không có dấu , phân cách
```

```python
## Sử dụng labels với series (labels KHÁC NHAU)
##  Tạo series từ list
s = pd.Series([9, 2, 7, 1, 6], index = ['P1', 'P2', 'P3', 'P4', 'P5'])
print('Các GIÁ TRỊ của s:', s.values)
print('Các CHỈ SỐ  của s:', s.index)

## Sử dụng labels với series (labels có thể TRÙNG NHAU !)
s = pd.Series([9, 2, 7, 1, 6], index = ['P1', 'P2', 'P3', 'P2', 'P5'])
print('Các GIÁ TRỊ của s:', s.values)
print('Các CHỈ SỐ  của s:', s.index)
```

```python
## Sử dụng dictionary (keys KHÁC NHAU)
## Tạo series từ dictionary
transactions = {'P1':9, 'P2':2, 'P3':7, 'P4':1, 'P5':6}
s            = pd.Series(transactions)
print('Các GIÁ TRỊ của s:', s.values)
print('Các CHỈ SỐ  của s:', s.index)

## Sử dụng dictionary (keys TRÙNG NHAU --> sai sót do loại bỏ phần tử !)
transactions = {'P1':9, 'P2':2, 'P3':7, 'P2':1, 'P5':6}
s            = pd.Series(transactions)
print('Các GIÁ TRỊ của s:', s.values)
print('Các CHỈ SỐ  của s:', s.index)
```

##### *Thêm phần tử mới vào series (xem như dictionary)*

```python
## Sử dụng labels với series (labels KHÁC NHAU)
s       = pd.Series([9, 2, 7, 1, 6], index = ['P1', 'P2', 'P3', 'P4', 'P5'])
s['P6'] = 100 # new index = 'P6'
print(s)
```

##### Truy cập các phần tử của series: **indexing**, **fancy indexing**, **slicing** (implicit index), **masking**

```python
## Sử dụng các cơ chế
s = pd.Series([9, 2, 7, 1, 6])
print('Phần tử thứ 3:', s[2])                       # indexing
print('Phần tử thứ 2 và thứ 4:', s[[1, 3]].values)  # fancy indexing
print('Phần tử thứ 2 --> 4:', s[1:4].values)        # slicing
print('Phần tử > 6:', s[s > 6].values)              # masking
```

```python
## Sử dụng labels (labels KHÁC NHAU)
s = pd.Series([9, 2, 7, 1, 6], index = ['P2', 'P1', 'P3', 'P5', 'P4'])
print('Số lượng mặt hàng P3 =', s['P3'])
print('Số lượng mặt hàng P3, P4:', s[['P3', 'P4']].values) # fancy indexing
print('Số lượng mặt hàng P3-P4:', s['P3':'P4'].values)
print('Số lượng mặt hàng thứ 3 đến thứ 5:', s[2:5].values)

## Sử dụng labels (labels TRÙNG NHAU)
s = pd.Series([9, 2, 7, 1, 6], index = ['P2', 'P3', 'P3', 'P5', 'P4'])
print('\nSố lượng mặt hàng P3 =', s['P3'].values)
print('Số lượng mặt hàng P3, P4:', s[['P3', 'P4']].values)
print('Số lượng mặt hàng P3-P4:', s['P3':'P4'].values)
## 'P3' đầu tiên; nếu thứ tự labels trong dữ liệu 'P3', 'P4', 'P3' --> error
```

```python
## Sử dụng dictionary (keys KHÁC NHAU)
transactions = {'P2':9, 'P1':2, 'P3':7, 'P5':1, 'P4':6}
s = pd.Series(transactions)
print('Số lượng mặt hàng P3 =', s['P3'])
print('Số lượng mặt hàng P3, P4:', s[['P3', 'P4']].values)
print('Số lượng mặt hàng P3-P4:', s['P3':'P4'].values)
print('Số lượng mặt hàng thứ 3 đến thứ 5:', s[2:5].values)
```

```python
## Sử dụng loc --> EXPLICIT index   ##Có thể dùng cho chỉ số định nghĩa hoặc ngầm định
s = pd.Series([9, 2, 7, 1, 6], index = ['P2', 'P1', 'P3', 'P5', 'P4'])
print('Phần tử từ P1 đến P5 =', s.loc['P1':'P5'].values)

## Sử dụng iloc --> IMPLICIT index  ##Chỉ dùng cho chỉ số ngầm định
print('Phần tử từ 1 đến 3 =', s.iloc[1:4].values)
```

##### *Xóa phần tử của series*

```python
## Sử dụng chỉ số ngầm định (chỉ định bằng list, không dùng slice)
s = pd.Series([9, 2, 7, 1, 6])
print('Xóa phần tử 2 và 4:', s.values, '-->', s.drop([1,3]).values)

## Sử dụng labels
s = pd.Series([9, 2, 7, 1, 6], index = ['P2', 'P1', 'P3', 'P5', 'P4'])
print('Xóa phần tử 2 và 4:', s.values, '-->', s.drop(['P1', 'P5']).values)

## Xóa bỏ các phần tử có GIÁ TRỊ trùng lắp
s = pd.Series([9, 2, 7, 9, 6])
print('Xóa phần tử trùng lắp:', s.values, '-->', s.drop_duplicates().values)
```

##### *Sử dụng một số phương thức của series*

```python
s = pd.Series([9, 2, 7, 1, 6], index = ['P2', 'P1', 'P3', 'P5', 'P4'])
print('MAX            =', s.max())
print('MIN            =', s.min())
print('SUM            =', s.sum())
print('Cumulative SUM =', s.cumsum().values, '\n')
print(s.describe())
```

### **2. DataFrame**
---


*   Có cấu trúc tương tự ma trận 2 chiều ở bên numpy



```python
##------------------------------------------------------------------------------
## * Series    có thể được xem như 1 "cột" dữ liệu (đi kèm với 1 cột chỉ số)
## * Dataframe có thể được xem như nhiều series được ghép nối theo hàng dọc
##             hay là tập hợp các series CHIA SẺ cùng chỉ số
##             --> tạo ra bảng dữ liệu 2 chiều với các 'columns names'
##------------------------------------------------------------------------------
```

##### *Tạo dataframe*

```python
## Tạo dataframe từ 1 ma trận (2D-array)
matrix = np.random.rand(5, 3)
df     = pd.DataFrame(matrix)
print(df, '\n')

## Gán tên cột
df = pd.DataFrame(matrix, columns = ['f1', 'f2', 'f3'])
print(df)

## Gán tên cột và dòng
df = pd.DataFrame(matrix, index = ['u1', 'u2', 'u3', 'u4', 'u5'], columns = ['f1', 'f2', 'f3'])
print(df, '\n')
```

```python
## Tạo dataframe từ 1 series
mahh    = ['P1', 'P2', 'P3', 'P4', 'P5']
tenhh   = pd.Series(['Mặt hàng 1', 'Mặt hàng 2', 'Mặt hàng 3', 'Mặt hàng 4', 'Mặt hàng 5'])
soluong = pd.Series([9, 2, 7, 1, 6])

df    = pd.DataFrame(tenhh, columns = ['tenhh'])
print(df, '\n')

## Tạo dataframe từ nhiều series --> cặp dấu bao {}
df   = pd.DataFrame({'mahh' : mahh, 'tenhh' : tenhh, 'soluong' : soluong})
print(df)
```

```python
## Tạo dataframe từ dictionary --> chuyển sang series bằng pd.Series()
tenhh_dict   = {'P1' : 'Mặt hàng 1', 'P2' : 'Mặt hàng 2', 'P3' : 'Mặt hàng 3',
                'P4' : 'Mặt hàng 4', 'P5' : 'Mặt hàng 5'}
soluong_dict = {'P1' : 9, 'P2' : 2, 'P3' : 7, 'P4' : 1, 'P5' : 6}
df           = pd.DataFrame({'tenhh': pd.Series(tenhh_dict), 'soluong': pd.Series(soluong_dict)})
print(df)
```

```python
## Tạo dataframe từ 1 list
lst = [['Mặt hàng 1', 9], ['Mặt hàng 2', 2], ['Mặt hàng 3', 7], ['Mặt hàng 4', 1],
       ['Mặt hàng 5', 6]]
df  = pd.DataFrame(lst, columns = ['tenhh', 'soluong'], index = mahh)
print(df, '\n')

lst = [['PA1', 'Mặt hàng 1', 9], ['PA2', 'Mặt hàng 2', 2], ['PA3', 'Mặt hàng 3', 7],
       ['PA4', 'Mặt hàng 4', 1], ['PA5', 'Mặt hàng 5', 6]]
df  = pd.DataFrame(lst, columns = ['mahh', 'tenhh', 'soluong'])
print(df)
```

```python
## Thay thế giá trị của cột theo điều kiện
## df.loc[<condition>, <column name>] = <expression>
np.random.seed(1)
matrix = np.random.randint(0, 100, (10, 3))
df     = pd.DataFrame(matrix, columns = ['f1', 'f2', 'f3'])
print(df.to_string(index = False), '\n')

## Gán f3 = (f2 / 2) đối với những dòng có f2 là số chẵn
df.loc[df["f2"] % 2 == 0, "f3"] = df['f2'] // 2
print(df.to_string(index = False))
```

##### *Thao tác trên DÒNG*

```python
## Truy xuất tất cả các cột với dataframe[<row_idx]
df = pd.read_excel(folder + '/Data/Data.xls', sheet_name = 'Insurance')

print(df[1:2], '\n----------') # dòng 2
print(df[1:3], '\n----------') # dòng 2, 3

## Truy xuất chỉ một số cột dataframe[<row_idx>][<col_idx>]
print(df[1:2]['age'], '\n----------')
print(df[1:3][['age', 'bmi', 'smoker']])
```

```python
## Truy xuất dữ liệu theo dòng với dataframe.loc[<row_idx>, <col_idx>]
print('Các dòng từ chỉ số [1] đến chỉ số [3]')
print(df.loc[1:3], '\n----------')

print('Tuổi của dòng có chỉ số [1] là ', df.loc[1]['age'], '\n----------')
print(df.loc[1:3][['age', 'smoker']]) # lưu ý: PHẢI CÓ   [] đối với chỉ số cột
print(df.loc[1:3, 'age':'smoker'])    # lưu ý: KHÔNG CẦN [] đối với chỉ số cột
```

```python
## Truy xuất dữ liệu theo dòng với dataframe.iloc[<row_idx>, <col_idx>]
print('Các dòng từ chỉ số [1] đến chỉ số [3-1]')
print(df.iloc[1:3], '\n----------') # dòng 2, 3

print(df.iloc[1]['age'], '\n----------')
print(df.iloc[1:3][['age', 'smoker']]) # lưu ý: PHẢI CÓ [] đối với chỉ số cột
print(df.iloc[1:3], [[0, 5]])          # lưu ý: PHẢI CÓ [] đối với chỉ số cột
print(df.iloc[1:3, 0:5])               # lưu ý: không dùng slicing với LABELS
```
