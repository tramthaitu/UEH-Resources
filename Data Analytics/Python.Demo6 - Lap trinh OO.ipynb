#### **1. Định nghĩa class**
---


*   Khi định nghĩa tên:
    * attribute (đối với cả class attribute & instance attribute) thì phải để __name (để đảm bảo bên ngoài không thấy được thuộc tính)
    * methods thì không cần



```python
##==============================================================================
## Định nghĩa lớp Khoa
##==============================================================================
class Khoa:
    #---------------------------------------------------------------------------
    # CLASS ATTRIBUTE(S)
    #    - Nếu là super-class thì phải liệt kê đầy đủ tất cả attributes !
    #      --> sub-class sẽ kế thừa class attributes
    #---------------------------------------------------------------------------
    truong = 'UEH' # nếu là private (hidden) attribute thì tên bắt đầu bằng __


    #---------------------------------------------------------------------------
    # CONSTRUCTOR:
    #    - Khởi tạo object với INSTANCE ATTRIBUTE(S) mang những giá trị khác nhau
    #      đối với từng instance
    #    - Sub-class chỉ KẾ THỪA class attributes KHÔNG KẾ THỪA instance attributes
    #---------------------------------------------------------------------------
    def __init__(self, makh, tenkh):
        self.__makh   = makh
        self.__tenkh  = tenkh


    #---------------------------------------------------------------------------
    # Methods:
    # Nhóm set: thay đổi giá trị bên trong của attribute
    # Nhóm get: lấy giá trị bên trong của attribute
    #---------------------------------------------------------------------------
    def get_tenkh(self):
        return self.__tenkh


    #---------------------------------------------------------------------------
    # Dunder (double-underscore) methods: hidden methods
    #---------------------------------------------------------------------------
    def __eq__(self, other): ##Hàm nội tại
        return (self.__makh.__eq__(other.__makh))

    def __str__(self):
        return f'Khoa {self.__makh}: {self.__tenkh}'
```

```python
##==============================================================================
## Định nghĩa lớp GiangVien
##==============================================================================
class GiangVien:
    #---------------------------------------------------------------------------
    # CLASS ATTRIBUTE(S)
    #    - Nếu là super-class thì phải liệt kê đầy đủ tất cả attributes !
    #      --> sub-class sẽ kế thừa class attributes
    #---------------------------------------------------------------------------
    __magv      = ''
    __hogv      = ''
    __tengv     = ''
    __khoa      = ''
    __hoc_vi    = ''
    __gio_chuan = 0


    #---------------------------------------------------------------------------
    # CONSTRUCTOR:
    #    - Khởi tạo object với INSTANCE ATTRIBUTE(S) mang những giá trị khác nhau
    #      đối với từng instance
    #    - Sub-class chỉ KẾ THỪA class attributes KHÔNG KẾ THỪA instance attributes
    #---------------------------------------------------------------------------
    def __init__(self, maso, ho, ten, khoa):
        self.__magv  = maso
        self.__hogv  = ho
        self.__tengv = ten
        self.__khoa  = khoa


    #---------------------------------------------------------------------------
    # Methods
    #---------------------------------------------------------------------------
    def get_magv(self):
        return self.__magv

    def get_hoten(self):
        return self.__hogv + ' ' + self.__tengv

    def get_khoa(self):
        return self.__khoa

    def get_hoc_vi(self):
        return self.__hoc_vi

    def get_gio_chuan(self):
        return self.__gio_chuan

    def set_magv(self, magv):
        self.__magv = magv

    def set_hotengv(self, hogv, tengv):
        self.__hogv  = hogv
        self.__tengv = tengv

    def set_khoa(self, khoa):
        self.__khoa = khoa

    def set_hoc_vi(self, hoc_vi):
        self.__hoc_vi = hoc_vi

    def set_gio_chuan(self, so_gio):
        self.__gio_chuan = so_gio


    #---------------------------------------------------------------------------
    # Dunder (double-underscore) methods
    #---------------------------------------------------------------------------
    def __eq__(self, other):
        return (self.__magv.__eq__(other.__magv))

    def __str__(self):
        return f'Giảng viên {self.__magv}: {self.__hogv} {self.__tengv}'
```

```python
##==============================================================================
## Định nghĩa lớp con (SUB/CHILD class) BanChuNhiem
##==============================================================================
class BanChuNhiem(GiangVien): ## Lớp cha là GiangVien, sẽ được đặt trong dấu ( )
    #---------------------------------------------------------------------------
    # Class attributes
    #---------------------------------------------------------------------------
    __ngay_ket_thuc = ''


    #---------------------------------------------------------------------------
    # CONSTRUCTOR:
    #    - Khởi tạo object với INSTANCE ATTRIBUTE(S) mang những giá trị khác nhau
    #      đối với từng instance
    #---------------------------------------------------------------------------
    def __init__(self, maso, ho, ten, khoa, chuc_danh, ngay_bat_dau):
        self.__magv         = maso
        self.__hogv         = ho
        self.__tengv        = ten
        self.__khoa         = khoa
        self.__chuc_danh    = chuc_danh
        self.__ngay_bat_dau = ngay_bat_dau


    #---------------------------------------------------------------------------
    # Methods
    #---------------------------------------------------------------------------
    def get_chuc_danh(self):
        return [self.__chuc_danh, self.__ngay_bat_dau, self.__ngay_ket_thuc]

    def get_gio_chuan(self):
        return self.__gio_chuan


    def set_ngay_ket_thuc(self, ngay):
        self.__ngay_ket_thuc = ngay

    def set_gio_chuan(self, so_gio):
        self.__gio_chuan = so_gio * 0.8


    #---------------------------------------------------------------------------
    # Dunder (double-underscore) methods
    #---------------------------------------------------------------------------
    def __str__(self):
        txt = 'Trưởng khoa' if self.__chuc_danh == 'TK' else 'Phó khoa'
        return txt + f' {self.__khoa}: {self.__hogv} {self.__tengv}'
```

```python
##==============================================================================
## Định nghĩa lớp Sinh_vien
##==============================================================================
class SinhVien:
    #---------------------------------------------------------------------------
    # Class attributes (giá trị giống nhau đối với mọi instances được tạo lập)
    #---------------------------------------------------------------------------
    __hoc_bong        = 0
    __ngay_tot_nghiep = ''


    #---------------------------------------------------------------------------
    # CONSTRUCTOR:
    #    - Khởi tạo object với INSTANCE ATTRIBUTE(S) mang những giá trị khác nhau
    #      đối với từng instance
    #---------------------------------------------------------------------------
    def __init__(self, maso, ho, ten, khoa):
        self.__masv   = maso
        self.__hosv   = ho
        self.__tensv  = ten
        self.__khoasv = khoa


    #---------------------------------------------------------------------------
    # Methods
    #---------------------------------------------------------------------------
    def get_masv(self):
        return self.__masv

    def get_hoten(self):
        return self.__hosv + ' ' + self.__tensv

    def get_khoa(self):
        return self.__khoasv

    def get_hoc_bong(self):
        return self.__hoc_bong

    def set_hoc_bong(self, so_tien):
        self.__hoc_bong = so_tien


    #---------------------------------------------------------------------------
    # Dunder (double-underscore) methods
    #---------------------------------------------------------------------------
    def __eq__(self, other):
        return (self.__masv.__eq__(other.__masv))

    def __str__(self):
        return f'Sinh viên {self.__masv}: {self.__hosv} {self.__tensv}'
```
