***SUPPORT VECTOR MACHINE***
*   Phân lớp SVM tuyến tính (Linear SVM Classification)
  * Xây dựng siêu phẳng (*hyperplane*) có thể phân cách các lớp
  * Để tối ưu kết quả phân lớp thì phải xác định siêu phẳng (*hyperplane*) có **khoảng cách đến các điểm dữ liệu (*margin*) của tất cả các lớp xa nhất có thể**
      * Đường hyperplane tốt nhất -> Tạo biên cực đại & Cách đều hai lớp

*   Các bước:
  1. ***Super vector (Vector hỗ trợ)***: những phần tử
đứng sát biên, gần với đối phương nhất
  2. Kẻ được hai đường biên (Pháp tuyến với vector hỗ trợ)
  3. Kẻ đường trung trực với hai đường biên đó -> Hyperplane

*  Phân lớp SVM tuyến tính:
  * *hard margin:* phân tách tuyến tính -> nhạy cảm với outliers
  * *soft margin:* cân đối giữa độ rộng của biên cực đại và giới hạn số lượng quan sát đã vi phạm đường biên


**NONLINEAR SVM CLASSIFICATION**
*   Dữ liệu không thể phân cách tuyến tính *-> Tạo siêu phẳng bằng cách phi tuyến*
*   Thông thường chỉ dùng hàm đa thức
*  Giải pháp bổ sung features *-> Phân cách phi tuyến*

```python
## Thư viện
import joblib            as jlb
import matplotlib.pyplot as plt
import numpy             as np
import pandas            as pd
import seaborn           as sbn

from sklearn.model_selection import train_test_split
from sklearn                 import svm
```

---
## 1. **Chẩn đoán bệnh ung thư**
---

```python
## Dữ liệu Breast Cancer
data = pd.read_csv('/content/gdrive/My Drive/Colab Notebooks/ML/Supervised Learning/Data/Breast Cancer WI.csv')
print(data.head())
```

```python
## Mô tả dữ liệu: Biến phân lớp (target variable): 'Class'
y = data['Class']

## Danh sách các features
nb_features = data.shape[1] - 1
features    = data.columns[:nb_features]
X           = data[features]
print('* Số lượng features = %2d' %nb_features)
print('  Các features:', ', '.join(features))
```

```python
## Kích thước các lớp
print("Số trường hợp CÓ    bệnh (Malignant):", y[y == 'Malignant'].size)
print("Số trường hợp KHÔNG bệnh (Benign)   :", y[y == 'Benign'].size)
```

```python
## Chia tập dữ liệu thành training, test sets theo tỷ lệ 80:20
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .2, random_state = 1)
```

```python
##------------------------------------------------------------------------------
## Xây dựng mô hình
##------------------------------------------------------------------------------
model = svm.SVC(kernel = 'linear')
model.fit(X_train, y_train)
```

```python
## Predict
model.predict([X_test.iloc[0]])
```

```python
## get support vectors
model.support_vectors_
```

```python
## get indices of support vectors
model.support_
```

```python
## get number of support vectors from each class
model.n_support_
```

```python
## Lưu trữ mô hình để khai thác về sau
jlb.dump(model,'/content/gdrive/My Drive/Colab Notebooks/ML/Supervised Learning/Data/Breast Cancer.svm')
```

```python
## Khai thác mô hình đã được huấn luyện để đoán (Test set)
model  = jlb.load('/content/gdrive/My Drive/Colab Notebooks/ML/Supervised Learning/Data/Breast Cancer.svm')

tiep = 'C'
while (tiep.upper() == 'C'):
    idx    = int(eval(input(f'Nhập chỉ số dòng từ 0 đến {X_test.shape[0]}: ')))
    idx    = idx % X_test.shape[0]
    X_new  = X_test[idx:idx + 1]
    y_pred = model.predict(X_new)
    print('Mẫu', X_new.to_string())
    print('Mẫu được dự đoán là', 'CÓ bệnh' if y_pred == 'Malignant' else 'KHÔNG có bệnh')
    tiep = input('Tiếp tục (C/K) ? ')
```
