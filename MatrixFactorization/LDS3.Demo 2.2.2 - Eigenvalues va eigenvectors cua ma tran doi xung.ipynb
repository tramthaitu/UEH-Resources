```python
import numpy      as np
import warnings
warnings.filterwarnings('ignore')

from numpy.linalg import eig, inv
from numpy        import diag
```

```python
## Tạo ma trận A(5 x 5) chứa các giá trị ngẫu nhiên trong khoảng 1 - 10
m, n, min, max = 5, 5, 1, 10
A = np.random.randint(min, max + 1, (m, n))
print('Ma trận A', A.shape, ':\n', A)
```

```python
## Tạo ma trận đối xứng  P = A @ A.T
##Nếu có ma trận đối xứng thì trị riêng (nghiệm) luôn luôn có nghiệm thực
P = A @ A.T
P
```

```python
## Tìm eigenvalues và eigenvectors của P
values, vectors = eig(P)
```

```python
## Hiển thị kết quả: Trị riêng lambda(r: số thực, i: số phức)
for j in range(len(values)):
    eigenvalue = values[j]

    r = values[j].real
    i = values[j].imag
    print('Eigenvalue[%1d]' %j, '= (%f,' %r, '%f)' %i)

print('\nEigenvectors', vectors.shape, ':\n', vectors)
```

---
#### **Nhận xét:**
- Ma trận ĐỐI XỨNG có các giá trị riêng đều là SỐ THỰC !
---

```python
## Tái tạo P từ các eigenvectors và eigenvalues: P = B @ Lambda @ B_1
from numpy.linalg import inv
B      = vectors
Lambda = diag(values)

print('Tái tạo ma trận P:\n', np.rint(B @ Lambda @ inv(B)).astype(int))
```
