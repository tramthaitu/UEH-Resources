## **BT thực hành 2.1.4: QR Decomposition**
- Tạo ma trận A(4x5) chứa các giá trị ngẫu nhiên trong khoảng 5 - 10
- Phân tích thành các thành phần Q, R. In kết quả
- Tái tạo ma trận A từ các thành phần Q, R.


> <u>Cập nhật</u>: **10/2023**



```python
## Thư viện
import numpy        as np
import warnings
warnings.filterwarnings('ignore')

from   numpy.linalg import qr
```

```python
## Tạo ma trận A(4, 5) với các giá trị ngẫu nhiên từ 1 đến 10
m, n, min, max = 4, 5, 1, 10
A = np.random.randint(min, max + 1, (m, n))
print('Ma trận A(', A.shape, ':\n', A)
```

```python
## Phân rã Q, R
## Các giá trị tham số
##     mode = 'reduced', 'complete': Q, R with dimensions (m, m), (m, n) (default)
Q, R = qr(A)
```

```python
## Ma trận vuông trực chuẩn Q
print('Ma trận Q(', Q.shape, ':\n', Q)
```

```python
## Kiểm chứng tính trực giao của Q
print(np.rint(Q @ Q.T).astype(int)) # rint(): làm tròn số
```

```python
## Ma trận tam giác trên R
print('Ma trận R(', R.shape, ':\n', R)
```

```python
## Tái tạo A từ các thành phần Q, R (kiểm chứng lại phép phân rã)
print(np.rint(Q @ R).astype(int))
```
